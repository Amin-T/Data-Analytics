```python
"""
Manipulation of csv files with pandas

In this file, I uses pandas datafram for manipulationg a csv file, which is a report obtained from content analysis of
LTSs of Islamic banks in 16 countries during years 2010-2019 using Yoshikoder application (http://yoshikoder.sourceforge.net/).

Provided by: Amin
"""

import pandas as pd
import re

Yoshi_report = pd.read_csv('Content_report-utf8.csv')
Yoshi_report.rename(columns={'Unnamed: 0': 'Index'}, inplace = True)
Yoshi_report.head()
```

```python
Bank = []
Country = []
Year =[]
Message = []
for i in Yoshi_report.index:
    splits = re.split('-|_|.txt', Yoshi_report.loc[i, 'Index'])
    Bank.append(splits[1])
    Country.append(splits[0])
    Year.append(int(splits[3]))
    Message.append(splits[2])
    
Yoshi_report.insert(0, 'Bank', Bank, True)
Yoshi_report.insert(1, 'Country', Country, True)
Yoshi_report.insert(2, 'Year', Year, True)
Yoshi_report.insert(3, 'Message', Message, True)
Yoshi_report.drop('Index', axis=1, inplace = True)
Yoshi_report.head()
```
