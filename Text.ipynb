```python
"""
The codes performe simple pre-proccesing of multiples .txt files

Provided by: Amin
"""

#pip install nltk
#nltk.download('stopwords')
#nltk.download('wordnet')

import glob
import os
import string

# Read .txt files
os.chdir('C:\\Users\\Amin\\Desktop\\Master Thesis\\Data\\Content_Analysis\\files1')
Data = {}
for file in glob.glob('*.txt'):
    with open(file, 'r', encoding= 'utf-8') as f:
        Data[file.split('.')[0]] = f.read()
```

```python
names = Data.keys()
Data_prep = {}
punc = set(string.punctuation)
from nltk.corpus import stopwords
stop = stopwords.words('english')
for i in names:
    lower = Data[i].lower()
    num_free = ''.join(x for x in lower if not x.isdigit())
    punc_free = ''.join(x for x in num_free if x not in punc)
    stop_free = ' '.join(x for x in punc_free.split() if x not in stop)
    Data_prep[i] = stop_free
print(Data_prep[i])
```

```python
path = 'C:\\Users\\Amin\\Desktop\\Master Thesis\\Data\\Content_Analysis\\files2'
for i in names:
    with open(path+'\\'+i+'.txt', 'w', encoding= 'utf-8') as f:
        f.write(Data_prep[i])
```
